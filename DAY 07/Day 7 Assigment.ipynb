**Encoding**

Our objective was to find out the factors which need to be control to curb the attrition of employee's, as the Attrition is a categorical variable it's needed to be encoded so as to proceed with further analysis,here two such encoding are employed,namely-

**1)** Label encoding

**2)** One-Hot encoding

**Label encoding**

Label encoding simply assign numeric values to different categories starting from 0, here we are encoding only Attrition field where **'Yes'=1** and **'No'=0.**

```python
from sklearn.preprocessing import LabelEncoder
encoder1 = LabelEncoder()
dataset2['Attrition'] = encoder1.fit_transform(dataset2['Attrition'])
```

**Obtained dataset after encoding of Attrition column**

```python
dataset2
```

**Corresponding Correlation Matrix when only attrition column is encoded**

```python
dataset2.corr()
```

**Droping columns with NaN(Not a number) values**

```python
dataset3=dataset2.drop(['StandardHours','EmployeeCount','Over18'],axis=1)
dataset3
```

**Corresponding Correlation Matrix after dropping NaN columns**

```python
dataset3.corr()
```

**Heatmap showing various correlation between different fields**

```python
fig, ax = plt.subplots(figsize=(20,20))
sns.heatmap(dataset3.corr(), ax=ax, annot=True, vmin=-1, vmax=1, center=0, cmap='coolwarm')
```

**Inference**

The above heat map shows that **Attrition** has **weak positive correlation** with **PercentSalaryHike** and **NumCompaniesWorked**, **strong negative correlation** with **Age, YearsAtCompany, YearsWithCurrManager** and **TotalWorkingYears**.

Moreover, these results are not accurate as we have not included other categorical columns such as BusinessTravel, Department, EducationField etc, so further we will encode these columns as well in order to obtain better results. 

**Label encoding of all the other categorical columns**

```python
from sklearn.preprocessing import LabelEncoder
encoder2 = LabelEncoder()
dataset3['BusinessTravel'] = encoder2.fit_transform(dataset3['BusinessTravel'])

#encodes 'Sales'=2,'Research & Development'=1,'Human Resources'=0 for Department column
from sklearn.preprocessing import LabelEncoder
encoder3 = LabelEncoder()
dataset3['Department'] = encoder3.fit_transform(dataset3['Department'])

#encodes 'Technical Degree'=5,'Other'=4,'Medical'=3,'Marketing'=2,'Life Sciences'=1,'Human Resources'=0 for EducationField column
from sklearn.preprocessing import LabelEncoder
encoder4 = LabelEncoder()
dataset3['EducationField'] = encoder4.fit_transform(dataset3['EducationField'])

#encodes 'Male'=1,'Female'=0 for Gender column
from sklearn.preprocessing import LabelEncoder
encoder5 = LabelEncoder()
dataset3['Gender'] = encoder5.fit_transform(dataset3['Gender'])

#encodes 'Sales Representative'=8,'Sales Executive'=7,'Research Scientist'=6,'Research Director'=5,'Manufacturing Director'=4,'Manager'=3,'Laboratory Technician'=2,'Human Resources'=1,'Healthcare Representative'=0 in JobRole column
from sklearn.preprocessing import LabelEncoder
encoder6 = LabelEncoder()
dataset3['JobRole'] = encoder6.fit_transform(dataset3['JobRole'])

#encodes 'Single'=2,'Married'=1,'Divorced'=0 in MaritalStatus column
from sklearn.preprocessing import LabelEncoder
encoder7 = LabelEncoder()
dataset3['MaritalStatus'] = encoder7.fit_transform(dataset3['MaritalStatus'])
```

**Corresponding dataset after all the categorical columns are encoded**

```python
dataset3
```

**Corresponding correlation matrix**

```python
dataset3.corr()
```

**Heatmap showing various correlation between different fields**

```python
fig, ax = plt.subplots(figsize=(20,20))
sns.heatmap(dataset3.corr(), ax=ax, annot=True, vmin=-1, vmax=1, center=0, cmap='coolwarm')
```

**Inference**

The above heat map shows that **Attrition** has **weak positive correlation** with **PercentSalaryHike, NumCompaniesWorked** and a **strong positive correlation** with **MaritalStatus**. Also,**strong negative correlation** of Attrition with **Age, YearsAtCompany, YearsWithCurrManager** and **TotalWorkingYears** exist.

Moreover, these results shows the correlation and dependency of Attrition with other fields as a whole but our objective was to sort out and check the dependency of various fields with Attrition='Yes', for this we have to perform one-hot encoding.

**Copying the original dataset for one-hot encoding**

```python
dataset_new=dataset.copy()
dataset_new
```

**One-hot encoding**

It is the most popular encoding that is opted for categorical columns, basically it introduces seperate columns for each category in the dataframe and assign values to all the records as 1 when record falls in that particular category and else as 0.

**One-hot encoding of all the categorical columns**

```python
df_dummy=pd.get_dummies(dataset_new,columns=['Attrition','BusinessTravel','Department','EducationField','Gender','JobRole','MaritalStatus','Over18'])
```

**Dataframe after one-hot encoding**

```python
df_dummy
```

**Displays the complete information about the encoded dataset.**

```python
df_dummy.info()
```

**Droping columns that have constant value**

```python
dataset4=df_dummy.drop(['StandardHours','EmployeeCount','Over18_Y'],axis=1)
dataset4
```

**Scaling**

Since, the dataset has a wide range of values in each column, thus scaling is needed to be performed so as to compress each data in a range of 0 to 1 and to ensure this a **MinMaxScaler** is used. 

**MinMax Scaling**

```python
from sklearn import preprocessing
min_max_scaler=preprocessing.MinMaxScaler()
dataset5=pd.DataFrame(min_max_scaler.fit_transform(dataset4))
```

**After MinMax scaling the dataframe loses it's column names, so here we will regenerate the column names using the original dataframe.**

```python
dataset5.columns = dataset4.columns
```

**Corresponding obtained dataset**

```python
dataset5
```

**Corresponding correlation matrix**

```python
dataset5.corr()
```

**Heatmap showing various correlation between different fields**

```python
fig, ax = plt.subplots(figsize=(40,40))
sns.heatmap(dataset5.corr(), ax=ax, annot=True, vmin=-1, vmax=1, center=0, cmap='coolwarm')
```

**Final Inferences**

**When Attrition is 'Yes'-**

**1) Strong positive correlation** exist for **MaritalStatus_Single** which means that employees who are single leaves the XYZ company more often.

**2) Strong positive correlation** exist for **BusinessTravel_Travel_Frequently** which means that employees who are made to travel frequently leaves the XYZ company more often.

**3) Positive correlation** exist for **EducationField_Human Resources** which means employee with educational background in Human Resources tends to leave XYZ company more often.

**4)  Positive correlation** exist for **Department_Human Resources** which means employee in Human Resources department tends to leave XYZ company more often, also same can be inferred from point 3 as well.

**5)  Strong negative correlation** exist for **TotalWorkingYears** which means that employees who have less number of total working years or experience leaves the XYZ company more often.

**6) Strong negative** and **equal correlation** exist for **Age** and **YearsWithCurrManager** which means that younger employees and employees with less experience with current manager tend to leave the XYZ company more often.


**7)  Strong negative correlation** exist for **YearsAtCompany** which means that employees who have spent less number of  years in company leaves the XYZ company more often.

**Conclusion**

From the above obtained inferences based on the final heatmap it's clear that Company XYZ lags behind in serving their new employees and must draft some good policies for the employees that are young,unmarried and have just joined the aforesaid organization.

Apart from that the main focus of the company should be on the Human Resource department as this department contributes for maximum attrition of the employees, discussion could be done with the employees at all levels of the department so as to understand and rectify the problems they are facing in a better way.

Employees with frequent business travel history must be taken care of either by giving them some sort of emoluments or a means of good travel comforts needed to be provided as per their demands.

So, Attrition in XYZ Company can be controlled by following above mentioned small measures and that may lead to the overall growth of the organization.

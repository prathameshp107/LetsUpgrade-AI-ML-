```python
import pandas as pd
dataset=pd.read_csv("general_data.csv")
dataset.head()
```

```python
dataset.info()
```

```python
dataset1=dataset.dropna()
dataset1
```

```python
dataset2=dataset1.drop_duplicates()
dataset2
```

```python
dataset3=dataset2[dataset2['Attrition']=='Yes']
dataset3
```

```python
dataset4=dataset2[dataset2['Attrition']=='No']
dataset4
```

**Problem Statement 1**

**H0:** There is no significant differnce in the employee's age between Attrition(Yes) and Attrition(No).(Null hypothesis)

**H1:** There is significant differnce in the employee's age between Attrition(Yes) and Attrition(No).(Alternate hypothesis)

Attrition can take up value as either "Yes" or "No", this give us two samples-

**1st Sample-** Employee with Attrition as "Yes" with certain age values.

**2nd Sample-** Employee with Attrition as "No" with certain age values.

Both these samples are independent so we have to apply Mann Whitney Test

```python
from scipy.stats import mannwhitneyu
stats1,p1=mannwhitneyu(dataset3.Age,dataset4.Age)
print("Critical value:",stats1)
print("p value:",p1)
```

**Conclusion**

Clearly, the **p value<0.05** so null hypothesis is rejected and alternate hypothesis is accepted i.e **There is significant differnce in the employee's age between Attrition(Yes)and Attrition(No).**
